#  Data Pro-cessing

```python
def create_data_frame(data):
    img_id_list = []
    img_path_list = []
    caption_1_list = []
    caption_2_list = []
    question_list = []
    answer_list = []
    topic_list = []
    for row_id in range(len(data)):
        img_id, img_path, caption1, caption2, id, question_answer = data.iloc[row_id]
        
        questions = question_answer["questions"]
        answers = question_answer["answers"]
        topics = question_answer["topics"]
        
        img_id_list += [img_id]*len(questions)
        img_path_list += [img_path]*len(questions)
        caption_1_list += [caption1]*len(questions)
        caption_2_list += [caption2]*len(questions)
        question_list += questions
        answer_list += answers
        topic_list += topics

    dataframe = {
        "Img id": img_id_list,
        "Img path": img_path_list,
        "Caption 1": caption_1_list,
        "Caption 2":caption_2_list,
        "Question":question_list,
        "Answer": answer_list,
        "Topic":topic_list
    }
    cartoon_data = pd.DataFrame(dataframe)
    cartoon_data['id'] = cartoon_data.index
    cartoon_data = cartoon_data[['id', "Img id", 'Img path', 'Caption 1', 'Caption 2', 'Question', 'Topic','Answer']]
    return cartoon_data
```

```python
cartoon_VQA = create_data_frame(df)
```

```python
cartoon_VQA.to_csv("raw_data.csv", index = False)
```
